```python
import requests
import json
import os
import pandas as pd
pd.set_option('display.max_columns', 30)
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
# query data of 1 day of taxi trips two months ago

current_datetime = datetime.now() - relativedelta(months=2) # 2 months earlier from now
formatted_datetime = current_datetime.strftime("%Y-%m-%d") # formating the 

url = (
    f"https://data.cityofchicago.org/resource/ajtu-isnz.json?"
    f"$where=trip_start_timestamp >= '{formatted_datetime}T00:00:00' "
    f"AND trip_start_timestamp <= '{formatted_datetime}T23:59:59'&$limit=30000"
)

headers = {'X-App-Token' : os.environ.get('CHICAGO_API_TOKEN')}

response = requests.get(url)

data = response.json()
```

## Transformation: create helper column for weather

```python
taxi_trips["trip_start_timestamp"] = pd.to_datetime(taxi_trips["trip_start_timestamp"]) # first -> convert to datetime dtype
taxi_trips["datetime_for_weather"] = taxi_trips["trip_start_timestamp"].dt.floor("H")
```

## Check joining the trips and the weather data

```python
# Get weather data and create a dataframe out of it

current_datetime = datetime.now() - relativedelta(months=2)
formatted_datetime = current_datetime.strftime("%Y-%m-%d")

url = "https://archive-api.open-meteo.com/v1/era5"

params = {
    "latitude": 41.85,
    "longitude": -87.65,
    "start_date": formatted_datetime,
    "end_date": formatted_datetime,
    "hourly": "temperature_2m,wind_speed_10m,rain,precipitation"
}

response = requests.get(url, params = params)

weather_data = response.json()

weather_data_filtered = {
    "datetime": weather_data["hourly"]["time"],
    "tempretaure": weather_data["hourly"]["temperature_2m"],
    "wind_speed": weather_data["hourly"]["wind_speed_10m"],
    "rain": weather_data["hourly"]["rain"],
    "precipitation": weather_data["hourly"]["precipitation"],
}

weather_df = pd.DataFrame(weather_data_filtered)

weather_df["datetime"] = pd.to_datetime(weather_df["datetime"])

weather_df.head()
```

```python
taxi_trips_with_weather = taxi_trips.merge(weather_df, left_on = 'datetime_for_weather', right_on = 'datetime')

taxi_trips_with_weather.head()
```

## Data modeling

```python
# create dimension table (Series) for payment type ID-s
payment_type_master = taxi_trips['payment_type'].drop_duplicates().reset_index(drop = True)
```

```python
# create dimension table (DataFrame) for payment type ID-s
payment_type_master = pd.DataFrame(
    {
        'payment_type_id' : range(1, len(payment_type_master) + 1),
        'payment_type' : payment_type_master
    }
)

payment_type_master
```

```python
# create dimension table (Series) for company ID-s
company_master = taxi_trips['company'].drop_duplicates().reset_index(drop = True)
```

```python
# create dimension table (DataFrame) for company ID-s
company_master = pd.DataFrame(
    {
        'company_id' : range(1, len(company_master) + 1),
        'company' : company_master
    }
)

company_master
```

```python
# write out two columns of taxi trips DataFrme which are get ID-s
taxi_trips[['payment_type', 'company']]
```

```python
# taxi trips DataFrame memory usage with full name of payment types and company names
taxi_trips.memory_usage(deep = True).sum()
```

```python
# create a new DataFrame for taxi trips replaced payment types and company names with ID-s
taxi_trips_id = taxi_trips.merge(payment_type_master, on = 'payment_type')

taxi_trips_id = taxi_trips_id.merge(company_master, on = 'company')

taxi_trips_id.drop(['payment_type', 'company'], axis = 1, inplace = True)

```

```python
# taxi trips DataFrame memory usage with ID-s of payment types and company names
taxi_trips_id.memory_usage(deep = True).sum()
```
